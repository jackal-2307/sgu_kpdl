### So sánh train và test

```python
# cài nếu chưa có
!pip install sweetviz --quiet

import pandas as pd
import sweetviz as sv

# đọc dữ liệu
df = pd.read_csv("marketing_campaign.csv")
# tối ưu: nếu file quá lớn, lấy mẫu
# df = df.sample(5000, random_state=1)

# 1) Phân tích một dataframe
report = sv.analyze(df)          # hoặc sv.analyze([df, "Target"]) nếu bạn chỉ rõ biến mục tiêu
report.show_html("sweetviz_report.html")  # lưu file HTML

# 2) So sánh train vs test
train = df.sample(frac=0.7, random_state=1)
test = df.drop(train.index)
compare_report = sv.compare([train, "Train"], [test, "Test"])
compare_report.show_html("sweetviz_compare.html")
```

## test WordCloud

**WordCloud (đám mây từ)** là một hình trực quan hiển thị tần suất xuất hiện của từ trong một tập văn bản.

Từ xuất hiện nhiều sẽ được vẽ lớn hơn; từ ít xuất hiện thì nhỏ hơn.

Rất hữu ích để nhìn nhanh chủ đề/keyword trong dữ liệu text (ví dụ feedback, mô tả, cột Education, Marital_Status…).

```python
%matplotlib inline
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = """Data Science is an interdisciplinary field that uses statistics,
machine learning, and programming skills to extract knowledge from data."""

wc = WordCloud(width=800, height=400, background_color='white').generate(text)

plt.figure(figsize=(10,4))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud Example")
plt.show()
# Lưu file để chắc chắn
wc.to_file("test wc.png")
print("Đã lưu: test wc.png")
```

```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

df = pd.read_csv("marketing_campaign.csv")
# clean column names
df.columns = [c.replace('\ufeff','').strip() for c in df.columns]

# Tìm cột 'education' (case-insensitive)
target = None
for c in df.columns:
    if c.lower().strip() == 'education':
        target = c
        break
# nếu không tìm chính xác, thử tìm chứa 'education'
if target is None:
    for c in df.columns:
        if 'education' in c.lower():
            target = c
            break

# Nếu vẫn không tìm thấy, fallback: lấy tất cả cột object/text
if target:
    print("Sử dụng cột:", target)
    text_data = " ".join(df[target].astype(str).dropna().values)
else:
    print("Không tìm thấy cột 'Education'. Sử dụng toàn bộ các cột text để tạo WordCloud.")
    text_cols = df.select_dtypes(include=['object']).columns.tolist()
    print("Các cột text dùng:", text_cols)
    if not text_cols:
        raise ValueError("Không có cột text nào để tạo WordCloud.")
    # ghép tất cả text (loại NaN)
    text_data = " ".join(
        (" ".join(row) for row in df[text_cols].fillna("").astype(str).values)
    )

# Tạo WordCloud
wc = WordCloud(width=1000, height=500, background_color='white', collocations=False).generate(text_data)

# Hiển thị trong notebook / VS Code notebook
plt.figure(figsize=(12,6))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.title("WordCloud từ marketing_campaign.csv", fontsize=16)
plt.show()

# Lưu file để chắc chắn
wc.to_file("wordcloud_marketing.png")
print("Đã lưu: wordcloud_marketing.png")
```
