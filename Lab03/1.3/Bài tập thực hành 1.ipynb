```python
# ...existing code...
# Cài đặt thư viện nếu cần và tạo báo cáo AutoViz
import sys, subprocess, os
try:
    from autoviz.AutoViz_Class import AutoViz_Class
except Exception:
    subprocess.run([sys.executable, "-m", "pip", "install", "autoviz", "--quiet"], check=False)
    from autoviz.AutoViz_Class import AutoViz_Class

import pandas as pd
os.makedirs("Reports", exist_ok=True)

df = pd.read_csv("marketing_campaign.csv")

AV = AutoViz_Class()
# Chạy AutoViz: các tham số có thể điều chỉnh (depVar = biến mục tiêu nếu có)
# chart_format: 'png' hoặc 'svg' (hình sẽ hiển thị inline trong notebook)
_ = AV.AutoViz(
    filename="",        # nếu muốn dùng file thay vì df, để đường dẫn ở đây
    sep=",",
    depVar="",
    dfte=df,            # truyền dataframe trực tiếp
    header=0,
    verbose=1,
    lowess=False,
    chart_format="png",
    max_rows_analyzed=20000,
    max_cols_analyzed=50
)

# Nếu muốn lưu toàn bộ notebook (gồm SweetViz + AutoViz output) sang HTML từ terminal:
# jupyter nbconvert --to html "Bài tập thực hành 1.ipynb" --output-dir Reports
print("Xong. Báo cáo AutoViz hiển thị trong notebook. Để lưu HTML, dùng nbconvert như chú thích.")
# ...existing code...
```
